```python
from scipy.stats import entropy
from collections import Counter
import numpy as np
from scipy.io import loadmat
from sympy.utilities.iterables import multiset_permutations
```

```python
path = 'mrc_data/subject50/subject50.mat'
data = np.array(loadmat(path)['X'])
data = data[::4, :306]
```

```python
w = 8
step = 4
perm_dict = {}
for i, a in enumerate(multiset_permutations(list(range(w)))):
    perm_dict[tuple(a)] = i
```

```python
data = data.reshape(-1)
np.random.shuffle(data.reshape(-1))
data = data.reshape(-1, 306)
```

```python
data = np.sin(np.arange(10000)/10) + np.random.randn(10000) * 2
```

```python
symbols = []
for i in range(data.shape[0] - w * step):
    vector = data[i:i+w*step, :]
    vector = np.argsort(vector[::step], axis=0)
    chs = [perm_dict[tuple(vector[:, ch])] for ch in range(306)]

    symbols.append(np.array(chs))
```

```python
symbols = np.array(symbols)
total = symbols.shape[0]
entropies = []
for ch in range(symbols.shape[1]):
    counts = Counter(symbols[:, ch])
    probs = np.array(list(counts.values()))
    entropies.append(entropy(probs/total))
```

```python
for e in entropies:
    print(e)
```
